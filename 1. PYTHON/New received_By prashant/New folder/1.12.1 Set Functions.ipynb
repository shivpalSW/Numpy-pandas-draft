## Intersection

```python
set1={'Pune','Chennai','Kanpur','Delhi'}
set2= {'Nagpur','Delhi','Pune','Banaras'}
set1.update(set2)
set1
```

```python
set1={'Pune','Chennai','Kanpur','Delhi'}
set2= {'Nagpur','Delhi','Pune','Banaras'}
set1=set1.intersection(set2)
set1
```

```python
set1={'Pune','Chennai','Kanpur','Delhi'}
set2= {'Nagpur','Delhi','Pune','Banaras'}
set3={'Delhi','Noida','Sangali'}
set4=set1.intersection(set2,set3)
set4
```
