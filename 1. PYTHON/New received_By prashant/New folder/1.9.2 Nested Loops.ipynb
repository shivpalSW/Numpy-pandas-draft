```python
num = eval(input('Enter any input: '))
if num<=1:
        print("Number out of prime range")
        
elif 1< num <=7:
    if num==2 or num==3 or num==5 or num==7:
        print("1: prime number")
    else: 
        print("1. Not a prime number")
            
else:
    if num>=8 and (num%2!=0 and num%3!=0 and num%5!=0 and num%7!=0):
        print("2: Prime number")
    else: 
        print("2. Not a prime number")
```

```python
number = int(input('Enter the number to check to prime'))

for i in range (2,number):
    if (number % i) == 0:
        print('Number is not prime')
        #print(i,"times",number//i,"is",number)
        break
else:
    print('Yes it is prime')
```
