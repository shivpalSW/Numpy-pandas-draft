```python
Difference between List & Tuple

List:
    1. []
    2. Mutable/ Changeble
    3. Ordered
    4. More functionalities than Tuple
    5. Less Memory efficient
    6. Slower than Tuple
    7. Duplicate values are allowed
    8. for single item it need not be comma separated provided than it is enclosed in square brackets
    
Tuple:
    1. ()
    2. Immutable / non-ediatable
    3. Ordered
    4. Lesser functionalities compared to List
    5. More memory efficient
    6. Faster than List
    7. Duplicate values are allowed
    8. for single item it needs to be comma separated when enclosed in round brackets
```

```python
n=int(input('enter any number- '))
sum1=0
sum2=1
print(sum1)
print(sum2)

while sum1<=(n-2):
    fib=sum1+sum2
    sum1=sum2
    sum2=fib
   
    print(fib)
```

```python
n=int(input('enter any number- '))
sum1=1
sum2=1

print(sum1,end=', ')
print(sum2,end='')

counter1=1

while counter1<(n-1): 
    
    fib=sum1+sum2
    sum1=sum2
    sum2=fib
    
    
    print(',',fib, end='' )
    counter1+=1
```
