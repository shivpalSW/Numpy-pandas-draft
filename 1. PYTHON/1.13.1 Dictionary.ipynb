## sorting around values of dictionary

```python
marks={'Maths': 90, 'Physics': 91, 'Bio': 93, 'Che': 95, 'History': 89}
new_dict={}
temp={}

for key,val in marks.items():
    temp[val]=key
    
print('Temp dict at the end of 1st for loop \n',temp)

for key in sorted(temp.keys()):
    
    new_dict[key] = temp[key]
    
new_dict
```
