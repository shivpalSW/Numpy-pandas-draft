# Chi-Square Test

```python
Cat vs Cat
Gender >> Cancer Detection 
Gender >> Loan Status

Each independent variable vs dependent variable

Higher the values >> Feature is more important, can be selected for model training

Null Hypothesis      : Two variables are independent   -   Feature is not imp
Alternate Hypothesis : Two variables are dependent     -   Feature is imp
    
Chi-square >> High  >> Two variables are dependent (We can use it for model)
Chi-square >> Low   >> Two variables are independent

p_val > 0.05    >> Null Hypothesis is True (Two variables are independent)
p_val <= 0.05   >> Null Hypothesis is False (Alternate Hypothesis is True)

0.18 > 0.05 >> Null Hypothesis is True >> Two variables are independent
0.0002 <= 0.05  >> Alternate Hypothesis >> Two variables are dependent
```

# Hypothesis:

```python
Explaination on the basis of limited evidence

1. Null Hypothesis(H0)       - Feature is not imp
2. Alternate Hypothesis(H1)  - Feature is imp


Only Null hypothesis is tested
```

```python
Age : 30 
Null Hypothesis : Mean of age column 30 
Alternate Hypothesis: Mean of age column is not equal to 30 
    
Null Hypothesis : BJP will win election
Alternate       : BJP will not win election

Evidence: Opinion >> 10 questions 
    
p_val > 0.05   >> Null Hypothesis is True
p_val <= 0.05   >> Null Hypothesis is False (Alternate Hypothesis)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('titanic.csv')
df['Gender'].replace({'male':0,'female':1},inplace = True)
df['Age'].fillna(df['Age'].median(),inplace = True)
df1 = df[['Survived','Pclass','Gender','Age','SibSp','Fare']]
# df1.info()
x = df1.drop('Survived',axis = 1)
y = df1['Survived']
```

```python
from sklearn.feature_selection import SelectKBest,chi2

best_features = SelectKBest(score_func=chi2,k = 5)
best_features.fit(x,y)

df_chi2 = pd.DataFrame({'Feature':x.columns,"Chi_Square":best_features.scores_})
df_chi2.sort_values('Chi_Square',ascending=False)
```

```python
chi2_score, p_val = chi2(x,y)
df_chi2 = pd.DataFrame({'Feature':x.columns,"Chi_Square":chi2_score,'P_val':np.around(p_val,5)})
df_chi2.sort_values('P_val')
```
