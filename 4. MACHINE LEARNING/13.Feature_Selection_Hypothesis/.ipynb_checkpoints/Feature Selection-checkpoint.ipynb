# 3. Information Gain

```python
IG determines the reduction in entropy

Calculation the IG of each variables with respect to target variables
```

```python
df = pd.read_csv('wine.csv')
# df
from sklearn.feature_selection import mutual_info_classif
# df.info()
x = df.drop('Wine',axis = 1)
y = df['Wine']
array = mutual_info_classif(x,y)
s1 = pd.Series(array)
s1.index = x.columns
s1.sort_values(ascending=True).plot(kind = 'barh')
```

# 4. Fishers Score 

```python
It returns rank of variable on the fishers criteria
```

```python
from skfeature.function.similarity_based import fisher_score
array = fisher_score.fisher_score(x.values,y)
s1 = pd.Series(array)
s1.index = x.columns
s1.sort_values(ascending=True).plot(kind = 'barh')
```
