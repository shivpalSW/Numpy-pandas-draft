## Calculation of Information Gain:


```python
how is the root node decided? how does the best split happen at the root node? 

for the same, we calculate the Information gain of the entire data.
The feature having high information gain will be decided as the root node.
```

```python
I.G = E(S) - sum[Weighted Avg* Entropy of each feature]

where,
I.G >> Information gain
E(S)>> Overall Entropy

Entropy >> It is used to find the major of impurity of node.
I.G     >> It is measurment of change  in entropy
```

#### OUTLOOK

```python
overcast : 4Y & 0N
rainy: 3Y & 2N
Sunny: 2Y & 3N

Total Samples: 9Y & 5N
    
Entropy(Outlook==overcast) = 0
Entropy(Outlook==rainy) = 0.9709505944546686
Entropy(Outlook==sunny) = 0.9709505944546686

Overall Entropy E(S) = 0.9402859586706311

Weights:
    outcast = 4/14
    rainy = 5/14
    Sunny = 5/14

I.G = Overall Entropy - Summation (weights * entropies of indivisuals)
    = 0.9402859586706311 - ( (4/14)*0 + (5/14)*0.9709505944546686 + (5/14)*0.9709505944546686 )
   
    = 0.24674981977443933
```

```python
#formula for entropy
Entropy= -(Prob Y)*np.log2(Prob Y) - (Prob N)*log2(Prob N)
```

```python
#calculating entropy of overcast
-(4/4)*np.log2(4/4)-(0/4)*np.log2(0/4)
```

```python
#calculating entropy of rainy
-(3/5)*np.log2(3/5) - (2/5)*np.log2(2/5)
```

```python
#calculating entropy of sunny
- (2/5)*np.log2(2/5) - (3/5)*np.log2(3/5) 
```

```python
# Calculating Overall Entropy
- (9/14)*np.log2(9/14) - (5/14)*np.log2(5/14) 
```

```python
# Calculating I.G for Outlook
0.9402859586706311 - ( (4/14)*0 + (5/14)*0.9709505944546686 + (5/14)*0.9709505944546686 )
```
