```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
df['bore']=df['bore'].astype(float)
df['normalized-losses']=df['normalized-losses'].astype(float)
df['stroke']=df['stroke'].astype(float)
df['horsepower']=df['horsepower'].astype(float)
df['peak-rpm']=df['peak-rpm'].astype(float)
df['price']=df['price'].astype(float)
```

### Simplest Encoding: replacing the distinct values of the columns with relevant numbers

```python
df1['num-of-cylinders'].replace({'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8},inplace=True)
```

```python
df1['num-of-cylinders'].dtype
```

```python
y=df1['price']
x=df1.drop('price',axis='columns')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=2)
```

```python
decision_tree=DecisionTreeRegressor()
decision_tree.fit(x_train,y_train)
```

```python
## testing data r2 score
y_pred=decision_tree.predict(x_test)
print("mse is", mean_squared_error(y_test,y_pred))
print('r2 score is', r2_score(y_test,y_pred))
```

```python
## training data r2 score
y_pred_train=decision_tree.predict(x_train)
print("mse is", mean_squared_error(y_train,y_pred_train))
print('r2 score is', r2_score(y_train,y_pred_train))
```

## hyper parameter tuning

```python
dt_model=DecisionTreeRegressor(random_state=1)
dt_model.fit(x_train,y_train)
```

```python
# dt_model=DecisionTreeRegressor()
```

#### GridsearchCV tuning

```python
dt_model.get_params().keys()
```

```python
hyperparameters = {'criterion':["squared_error", "friedman_mse", "absolute_error","poisson"],
                    'max_depth':np.arange(4,10),
                  'min_samples_leaf':np.arange(2,20),
                  'min_samples_split':np.arange(2,10)}
gscv_dt_model=GridSearchCV(dt_model,hyperparameters,cv=5)
gscv_dt_model
```

```python
gscv_dt_model.fit(x_train,y_train)
```

```python
gscv_dt_model.best_params_
```

```python
y_pred=gscv_dt_model.predict(x_test)
r2_score(y_test,y_pred)
```

```python
y_pred_train=gscv_dt_model.predict(x_train)
r2_score(y_train,y_pred_train)
```
