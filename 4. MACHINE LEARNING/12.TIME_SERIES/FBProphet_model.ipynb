```python
df.columns = ['ds', 'y']
df.head()
```

```python
# import prophet
```

```python
from prophet import Prophet
```

```python
# from fbprophet import Prophet
```

```python
y_train = df[:380]
y_test = df[380:]
y_train
```

```python
# Initialize the Model
model_m = Prophet()
```

```python
model_m.fit(y_train)
```

```python
model_m.component_modes # to check which modes we cam use as regressor
```

```python
y_test
```

```python
# Creating future dates
future_dates_m = model_m.make_future_dataframe(periods=120,freq='M') # y-test(96)+future(24)=120
future_dates_m
```

```python
prediction_m = model_m.predict(future_dates_m)
```

```python
prediction_m[476:]
```

```python
# Plot the predicted projection
model_m.plot(prediction_m)
```

```python
# Visualize each components[Trends,Weekly]
model_m.plot_components(prediction_m)
```

```python
from fbprophet.diagnostics import cross_validation
df_cv_m = cross_validation(model_m,initial='730 days', period='180 days',horizon='365 days')
df_cv_m.head(15)
```

```python
df_cv_m.tail(40)
```

```python
from fbprophet.diagnostics import performance_metrics
df_p_m = performance_metrics(df_cv_m)
df_p_m.head()
```

```python
from fbprophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(df_cv_m, metric='rmse')
```
