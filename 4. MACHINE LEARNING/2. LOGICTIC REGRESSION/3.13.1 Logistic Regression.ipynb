```python
Logistic Regression:
    It is a supervised machine learning algorithm which uses linear model however here the targeted variable is a categorical value & the independent variables can be categorical or countinuos.
    It is a machine learning classification algorithms is used to predict the probability of the targeted variable to which class it may belong.
    
Targetaed variable >> Categorical/Discrete values.

It is called as Logistic regression bcz we use the concept of linearity or the linear equation & additionally we use the log function to find log of odds so is the name.


Classes in targeted variable:
    Binary Class:
        0/1
        Yes/No
        Pass/Fail
        True/False
        High/Low
        Male/Female
        Left/Right
        
    Multi class classifier:
        
        Ordinal classification:
            High, Medium, Low
            High, Higher, Highest
            Good, Better, Best
            Nano, Micro, Macro
            
        Unordinal Classification:
            Apple, Banana, Pineapple, Peach
            Male, Female, ransgender
            Red, Blue, Green
            
            
            OVR (One Verses Rest/ One Verses All)
            Red or not red
            Red or everything apart from Red
            Blue or not blue
            Greeen or not Green
```
