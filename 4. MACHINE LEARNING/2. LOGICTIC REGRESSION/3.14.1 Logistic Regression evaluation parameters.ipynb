```python
Overfitting:
    Training Accuracy >> High
    Testing accuracy >> Low
    
    LOW BIAS & HIGH VARIANCE
    
    
Underfitting:
    Training Accuracy >> Low
    Testing accuracy >> Low   
    
    HIGH BIAS & LOW VARIANCE
    
NOTE: Testing accurcay is always lesser than the training accuracy. Only in very exceptional cases this may be reversed!
    
Best fit:
    Training Accuracy >> High
    Testing accuracy >> High     
    
    LOW BIAS & LOW VARIANCE
    
Bias :
    How close the points to the line (Linear Regression) ||| diff btw actual points & predicted points
    
    High Bias: Model is not capturing the training data >> training accuracy >> LOW
    Low Bias : Model is capturing the training data >> training accuracy >> High
        
Variance:
    Spread of points over the plot. High varinace>> spread of points away from each other & from the mean point.
    
    Low Variance:
        Training Accuracy : High >> 90%
        Testing Accurcy : High  >> 89%
            
    High Variance:
      Case 1:  
        Training Accuracy: High >> 90%
        Testing Accuracy: Low >> 88%
        
      Case 2:
        Training Accuracy: Low >> 74%
        Testing Accuracy: High >> 75%
        
```

```python
How to avoid Overfitting:
    Dropping the features (60 >> 50)
    Treating Outliers
    Prunning (cutting the branches of tree)
    Cross Validation
    Regularisation
    Ensembling Techninques (Bagging & Boosting)
```

```python
How to avoid Underfitting:
    Treating Outliers
    Add Features
    Add Observation
    Use corelated features
    Null Values Treatment
    
```
